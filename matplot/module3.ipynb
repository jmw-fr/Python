```python
import pandas as pd

exam_data = pd.read_csv('datasets/exams.csv')
exam_data.head()
```

```python
exam_scores = exam_data[['math score', 'reading score', 'writing score']]
exam_scores.head()
```

```python
exam_scores.describe()
```

```python
exam_scores_array = np.array(exam_scores)

bp = plt.boxplot(exam_scores_array, patch_artist=True)

colors = ['blue', 'grey', 'lawngreen']

for i in range(len(bp['boxes'])):
    bp['boxes'][i].set(facecolor=colors[i])
    bp['caps'][2*i + 1].set(color=colors[i])

plt.xticks([1, 2, 3], ['Math', 'Reading', 'Writing'])
```

```python
vp = plt.violinplot(exam_scores_array)
plt.xticks([1, 2, 3], ['Math', 'Reading', 'Writing'])
```

```python
vp = plt.violinplot(exam_scores_array, showmedians=True)
plt.xticks([1, 2, 3], ['Math', 'Reading', 'Writing'])
```

```python
vp = plt.violinplot(exam_scores_array, showmedians=True, vert=False)
plt.yticks([1, 2, 3], ['Math', 'Reading', 'Writing'])
```

```python
vp
```

```python
vp = plt.violinplot(exam_scores_array, showmedians=True, vert=False)
plt.yticks([1, 2, 3], ['Math', 'Reading', 'Writing'])
colors = ['blue', 'grey', 'lawngreen']

for i in range(len(vp['bodies'])):
    vp['bodies'][i].set(facecolor=colors[i])

vp['cmaxes'].set(color='maroon')
vp['cmins'].set(color='black')
vp['cbars'].set(linestyle=':')
vp['cmedians'].set(linewidth=6)

plt.legend(handles = [vp['bodies'][0], vp['bodies'][1]], labels=['Math', 'Reading'], loc='upper left')
```

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

np_data= pd.read_csv('datasets/national_parks.csv')
np_data.head()
```

```python
x = np_data['Year']

y = np.vstack([np_data['Badlands'], 
               np_data['GrandCanyon'], 
               np_data['BryceCanyon']])

labels = ['Badlands', 
          'GrandCanyon', 
          'BryceCanyon']

colors = ['sandybrown', 
          'tomato', 
          'skyblue']

plt.stackplot(x, y, 
              labels=labels,
              colors=colors,  # type: ignore
              edgecolor='grey'
              )

plt.legend(loc='upper left')
```

```python
np_data[['Badlands', 
         'GrandCanyon',
         'BryceCanyon']] = np_data[['Badlands', 
                                    'GrandCanyon',
                                    'BryceCanyon']].diff()

np_data.head()
```

```python
plt.figure(figsize=(10,6))

plt.stem(np_data['Year'],
         np_data['Badlands'])

plt.title('Change in Number of Visitors')
```
