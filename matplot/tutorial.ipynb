```python
x = numpy.linspace(start=0, stop=10, num=100)

plt.plot(x, numpy.sin(x), color='orange', label="sin(x)")
plt.plot(x, numpy.cos(x), color='blue', label="cos(x)")

plt.xlabel('x', color='green')
plt.ylabel('sin(x)', color='green')

plt.tick_params(axis='both', labelcolor='blue')

plt.legend()
plt.title('Playing with plots')

plt.xlim(1, 10)
plt.ylim(-1, 1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()

x = np.linspace(0, 10, 100)

ax1 = fig.add_axes([0, 0.6, 1, 0.4])
ax1.set_ylabel('sin(x)', color='red')
ax1.plot(x, np.sin(x), color='red')

ax2 = fig.add_axes([0, 0, 0.8, 0.4])
ax2.set_ylabel("cos(x)", color='green')
ax2.plot(x, np.cos(x), color='green')
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(10,4))

x = np.linspace(0, 5, 100)

ax1 = fig.add_axes([0, 0, 1, 1])
ax1.set_ylabel('sin(x)', color='red')
ax1.plot(x, np.sin(x), color='red')

ax2 = fig.add_axes([0.5, 0.5, 0.4, 0.4])
ax2.set_ylabel("cos(x)", color='green')
ax2.plot(x, np.cos(x), color='green')
```
